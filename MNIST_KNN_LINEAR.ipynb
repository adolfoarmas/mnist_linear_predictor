### Using Linear regresion to predict a number value by knowing the upper side of a handwritten digit

 *Stack:* Numpy, Pandas, MatPlotLib, SciKitLearn
 
 *Dataset:* SciKitLearn's `load_digits` dataset
 
 *Estimators:* KNeighborsRegressor, LinearRegression

The project results include a comparisson between KNN and LinearRegression.

```python
import numpy as np
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mnist = load_digits()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    mnist.data, mnist.target, test_size=0.3, random_state=50
)  
```

```python
X_train_splitted = X_train[:, :n_pixels // 2]    #<----- pixels from o to 31 (higher half of the images)
X_test_splitted = X_test[:, :n_pixels // 2]      #<----- pixels from o to 31 (higher half of the images)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression

ESTIMATORS = {
    "K-nn": KNeighborsRegressor(),
    "Linear regression": LinearRegression(),
}
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score

y_test_predict = dict()
for name, estimator in ESTIMATORS.items():
    estimator.fit(X_train_splitted, y_train)
    y_test_predict[name] = estimator.predict(X_test_splitted)
    #METRICS
    mae = mean_absolute_error(y_test, y_test_predict[name])
    mse = mean_squared_error(y_test, y_test_predict[name])
    r2 = r2_score(y_test, y_test_predict[name])
    explained_variance = explained_variance_score(y_test, y_test_predict[name])
    print("___"*18)
    print(name)
    print("---"*18)
    print("Mean Absolute Error (MAE):", mae)
    print("Mean Squared Error (MSE):", mse)
    print("RÂ² (R2 Score):", r2)
    print("Explained Variance Score:", explained_variance)
print("___"*18)
```

```python
image_shape = (8,8) #<----- Dimage dimention
n_cols = 1 + len(ESTIMATORS)
```

```python
plt.figure(figsize=(2 * n_cols, 2 * n_digits)) #<------ image size
plt.suptitle("Predictions of the digit knowing the upper half of the image using KNN vs. Lineal Regresor", )                #<------ Subtitle

for i in range(n_digits):
    
    true_digit = X_test[i]
    true_value = y_test[i]
    sub = plt.subplot(n_digits, n_cols, i * n_cols + 1, title="True Digits" if i == 0 else "")
    sub.axis("off")
    sub.imshow(true_digit.reshape(image_shape), cmap=plt.cm.gray, interpolation="nearest")
    sub.text(0.5, -0.1, f"True: {true_value}", color='blue', ha='center', va='center', transform=sub.transAxes)

    # Loop through the estimators
    for j, est in enumerate(sorted(ESTIMATORS)):
        
        completed_digit = np.hstack(X_test_splitted[i])
        prediction = y_test_predict[est][i]

        if i:
            sub = plt.subplot(n_digits, n_cols, i * n_cols + 2 + j)
        else:
            sub = plt.subplot(n_digits, n_cols, i * n_cols + 2 + j, title=est)

        sub.axis("off")
        sub.imshow(completed_digit.reshape([-1, image_shape[0]]), cmap=plt.cm.gray, interpolation="nearest")

        sub.text(0.5, -0.1, f"Pred: {"%.2f" % prediction}", color='red', ha='center', va='center', transform=sub.transAxes)

plt.show()
```
